```python
#example 1
week = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
for x in week:
  print(x)
```

```python
#example 2 (the break statement)
week = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
for x in week:
  print(x)
  if x == "thursday":
    break
```

```python
#example 3 (the break statement)
week = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
for x in week:
  if x == "thursday":
    break
  print(x)
```

```python
#example 4 (looping through a string)
for x in "week":
  print(x)
```

```python
#The range() Function
for x in range(6):
  print(x)
for x in range(2,6):
  print(x)
```

```python
#Nested loops
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x, y)
```

```python
#The Break Statement
i = 1
while i < 6:
 print(i)
 if i ==3:
  break
 i += 1
```

```python
#The continue Statement
i = 0
while i < 6:
 i += 1
 if i ==3:
  continue
 print(i)
```
