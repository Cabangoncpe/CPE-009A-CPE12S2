Comparison Operators

```python
t = 2
u = 3

print(t==u) #Equal
print(t!=u) #Not Equal
print(t>u) #Greater Than
print(t<u) #Less Than
print(t>=u) #Greater Than or Equal To
print(t<=u) #Less Than or Equal To

```

Logical Operators

```python
t = 2

print (t < 5 and t < 10) #and
print (t < 5 or t < 4) #or
print (not(t < 5 and t <10)) #not
```

Bitwise Operators

```python
v = 2

print(v&1) # & (AND)
print(v|3) # | (OR)
print(v^1) # ^ (XOR)
print(~v) # ~ (NOT)
print(v<<2) #<< (left shift)
print(v>>1) #>> (right shift)
```
